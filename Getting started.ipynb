### Data Types
* No need to mention type before the variable
* A variable understands its type by the data assigned to it. 
* Based on the assignment below, variable 'a' becomes data of type integer, b float & c string

```python
a = 5
b = 1.1
c = 'abc'
```

#### Type of data can dynamically change based on type of data assigned to it, a string variable when assigned integer becomes an integer.
#### Because of this, variable reusability becomes quite good & thus, very less noice in code

```python
type(c)
```

```python
c = 1
type(c)
```
